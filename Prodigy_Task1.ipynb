```python
#Importing all essential python libraries for Data Visualization
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm # Import the colormap module
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
#Loading Dataset
#Skipping first 4 rows and
df = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_131993.csv",skiprows=4)
col_index = df.columns.get_loc("2023")
df = df.iloc[:, :col_index + 1]
```

```python
df.describe()
```

```python
# Filter out aggregate country groups
countries_to_exclude = [
    "Low & middle income","Middle income","IBRD only","Early-demographic dividend","Upper middle income", "High income", "Low income",
    "Lower middle income", "Late-demographic dividend", "Post-demographic dividend", "IDA & IBRD total", "IDA total", "IDA blend",
    "IDA only", "Euro area", "East Asia & Pacific (excluding high income)","East Asia & Pacific","Europe & Central Asia (excluding high income)",
    "Europe & Central Asia","Latin America & Caribbean (excluding high income)","Latin America & Caribbean","Middle East & North Africa",
    "Middle East & North Africa (excluding high income)","North America","South Asia","Sub-Saharan Africa (excluding high income)","Sub-Saharan Africa",
    "World","Arab World", "Central Europe and the Baltics","Caribbean small states","Central African Republic", "Channel Islands",
    "Developing small island states","East Asia & Pacific (IDA & IBRD countries)","Europe & Central Asia (IDA & IBRD countries)",
    "Fragile and conflict affected situations","Heavily indebted poor countries (HIPC)","High income: nonOECD","High income: OECD",
    "Latin America & the Caribbean (IDA & IBRD countries)","Least developed countries: UN classification","Middle East & North Africa (IDA & IBRD countries)",
    "Pacific island small states","Small states","South Asia (IDA & IBRD)","Sub-Saharan Africa (IDA & IBRD countries)",
    "Other small states","Pre-demographic dividend","European Union","Least Developed Countries","Less Developed Regions",
    "More Developed Regions","Land Locked Developing Countries","Small Island Developing States","Eastern and South-Eastern Asia",
    "Northern Africa and Western Asia","Oceania","Northern America and Europe","Eastern and Southern Africa","Middle Africa",
    "Western Africa","Western Asia","OECD members","Not classified"
]

# Apply the filter to the DataFrame
df = df[~df["Country Name"].isin(countries_to_exclude)]

```

```python
year_cols = [col for col in df.columns if col.isdigit()]
df['Total Population'] = df[year_cols].sum(axis=1, skipna=True)
df
```

 **1. Find the top 10 countries by total population**






```python
# Find the top 10 countries by total population
top_10_countries =  df.sort_values(by='Total Population', ascending=False).head(10)
colors = cm.plasma(np.linspace(0, 1, len(top_10_countries)))

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country Name'], top_10_countries['Total Population'], color=colors)
plt.xlabel('Country Name')
plt.ylabel('Total Population')
plt.title('Top 10 Countries by Total Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**2. India Population Growth (2015–2023)**

```python
# India population trend
india = df[df["Country Name"] == "India"]
plt.figure(figsize=(8, 4))

# Filter year_cols to include only years from 2015 to 2023
recent_year_cols = [col for col in year_cols if int(col) >= 2015]

# Use recent_year_cols for plotting
plt.plot(recent_year_cols, india[recent_year_cols].values.flatten(), label="India", color='green')
plt.title("India Population Growth (2015–2023)")
plt.xlabel("Year"); plt.ylabel("Population"); plt.grid(True)
plt.tight_layout(); plt.show()
```

**3. Population comparsion of India, USA & China Overtime**

```python
# India, China, USA comparison
countries = df[df["Country Name"].isin(["India", "China", "United States"])]
plt.figure(figsize=(10, 5))
for _, row in countries.iterrows():
    plt.plot(year_cols, row[year_cols], label=row["Country Name"])

plt.legend(); plt.title("Population Growth: India, China, USA")
plt.xlabel("Year"); plt.ylabel("Population"); plt.grid(False)
plt.xticks(rotation='vertical')
plt.tight_layout(); plt.show()
```

**4. Population comparsion of Three Countries in 2000,2010,2020**

```python
# 3 countries in 2000, 2010, 2020
selected_years = ["2000", "2010", "2020"]
data = df[df["Country Name"].isin(["India", "China", "United States"])]

data_plot = data.set_index("Country Name")[selected_years].T
data_plot.plot(kind="bar", figsize=(10, 5))
plt.title("Population Comparison (2000, 2010, 2020)")
plt.xlabel("Year"); plt.ylabel("Population")
plt.tight_layout(); plt.show()
```

**5. Distribution of country populations in 2023**

```python
# Distribution of country populations in 2023
plt.figure(figsize=(8, 4))
sns.histplot(df["2023"].dropna(), bins=30, kde=False, color='skyblue')
plt.title("Population Distribution in 2023")
plt.xlabel("Population"); plt.ylabel("Number of Countries")
plt.tight_layout(); plt.show()
```

**6. Share of world population (Top 5) in 2023**

```python
#  Share of world population (Top 5) in 2023
# Find the top 5 countries by population in 2023
top5 = df.sort_values(by='2023', ascending=False).head(5)

top5 = top5.set_index("Country Name")
plt.figure(figsize=(6, 6))
top5["2023"].plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 5 Countries by Population Share (2023)")
plt.ylabel(""); plt.tight_layout(); plt.show()
```

**7. 1960 vs 2023 population**

```python
# 1960 vs 2023 population
df_clean = df.dropna(subset=["1960", "2023"])
plt.figure(figsize=(7, 5))
plt.scatter(df_clean["1960"], df_clean["2023"], alpha=0.6)
plt.title("Population: 1960 vs 2023")
plt.xlabel("Population in 1960"); plt.ylabel("Population in 2023")
plt.grid(True); plt.tight_layout(); plt.show()
```

**8. Population Share of Top 5 Countries Over Time**

```python
# Population Share of Top 5 Countries Over Time

# Find the top 5 countries by total population
top_5_countries_list = df.sort_values(by='Total Population', ascending=False).head(5)['Country Name'].tolist()

# Filter the original DataFrame to include only these top 5 countries
df_top5 = df[df['Country Name'].isin(top_5_countries_list)]

# Select only the country name and the year columns
area_chart_data = df_top5[['Country Name'] + year_cols].set_index('Country Name').T # Transpose for plotting years on x-axis

# Replace potential NaN values with 0 for plotting
area_chart_data = area_chart_data.fillna(0)

# Create the stacked area chart
plt.figure(figsize=(12, 6))
area_chart_data.plot(kind='area', stacked=True, ax=plt.gca()) # Use ax=plt.gca() to plot on the current figure

plt.title("Population Share of Top 5 Countries Over Time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside plot
plt.tight_layout()
plt.show()
```
